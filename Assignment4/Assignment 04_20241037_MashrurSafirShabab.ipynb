```python
#TASK_1

class Calculator:
    
    def __init__ (self):
        print ("Let’s Calculate!")
        self.first_value = None
        self.operator = None
        self.second_value = None
    
    def add (self):
        x = self.first_value + self.second_value 
        print ("Result : ",x)
    
    def substract (self):
        x = self.first_value - self.second_value 
        print ("Result : ",x)
    
    def multiply (self):
        x = self.first_value * self.second_value 
        print ("Result : ",x)
    def divide (self):
        x = self.first_value / self.second_value 
        print ("Result : ",x)
    
#===============================================================================================================   

obj = Calculator()
obj.first_value = int(input("Value 1 : "))
obj.operator = input("Operator : ")
obj.second_value = int(input("Value 2 : "))

if obj.operator == "+":
    obj.add()
elif obj.operator == "-":
    obj.substract()
elif obj.operator == "*":
    obj.multiply()
elif obj.operator == "/":
    obj.divide()
```

```python
#TASK_3

class Panda:
    
    def __init__(self, name, gender, age):
        self.name = name
        self.gender = gender
        self.age = age
    
    def sleep (self,sleep=None):
        if sleep == None:
            print ( "{}'s duration is unknown thus should have only bamboo leaves".format(self.name))
        else:
            self.sleep = sleep
            if 5>=int(sleep)>=3:
                return ("{} sleeps {} hours daily and should have Mixed Veggies".format(self.name,sleep))
            elif 8>=int(sleep)>=6:
                return ("{} sleeps {} hours daily and should have Eggplant & Tofu".format(self.name,sleep))
            elif 11>=int(sleep)>=9:
                return ("{} sleeps {} hours daily and should have Broccoli Chicken".format(self.name,sleep))
        
#==============================================================================================================

panda1 = Panda("Kunfu","Male", 5) 
panda2=Panda("Pan Pan","Female",3) 
panda3=Panda("Ming Ming","Female",8)
print("{} is a {} Panda Bear who is {} years old".format(panda1.name,panda1.gender,panda1.age))
print("{} is a {} Panda Bear who is {} years old".format(panda2.name,panda2.gender,panda2.age))
print("{} is a {} Panda Bear who is {} years old".format(panda3.name,panda3.gender,panda3.age))
print("===========================") 
print(panda2.sleep(10)) 
print(panda1.sleep(4))
print(panda3.sleep())
```

```python
#Task_6

class Vehicle:
    
    def __init__ (self):
        self.x = 0
        self.y = 0
        
    def moveUp (self):
        self.y += 1
        
    def moveDown (self):
        self.y -= 1
        
    def moveLeft (self):
        self.x -= 1
        
    def moveRight (self):
        self.x += 1
        
    def print_position (self):
        print((self.x,self.y))
        

car = Vehicle() 
car.print_position()
car.moveUp()
car.print_position()
car.moveLeft()
car.print_position()
car.moveDown()
car.print_position()
car.moveRight()
```

```python
#Task_11

class Shinobi:
    
    def __init__ (self, name, rank, salary = 0, mission = 0):
        self.name = name
        self.rank = rank
        self.salary = salary
        self.mission = mission
    
    def calSalary(self,sal):
        self.mission = sal
        if self.rank == 'Genin':
            self.salary = self.mission * 50
        
        elif self.rank == 'Chunin':
            self.salary = self.mission * 100
            
        else:
            self.salary = self.mission * 500
            

    def printInfo(self):
        print ("Name:",self.name)
        print ("Rank:",self.rank)
        print ("Number of mission:",self.mission)
        print ("Salary:",self.salary)
    
    def changeRank(self,new_rank):
        self.rank = new_rank
        if self.rank == 'Genin':
            self.salaryy = self.mission * 50
        
        elif self.rank == 'Chunin':
            self.salaryy = self.mission * 100
            
        else:
            self.salaryy = self.mission * 500
        print ("Name:",self.name)
        print ("Rank:",self.rank)
        print ("Number of mission:",self.mission)
        print ("Salary:",self.salaryy)
        
naruto = Shinobi("Naruto", "Genin") 
naruto.calSalary(5)
naruto.printInfo() 
print('====================') 
shikamaru = Shinobi('Shikamaru', "Genin")
shikamaru.printInfo() 
shikamaru.changeRank("Chunin") 
shikamaru.calSalary(10) 
shikamaru.printInfo() 
print('====================')
neiji = Shinobi("Neiji", "Jonin") 
neiji.calSalary(5) 
neiji.printInfo()
```

```python
#Task_12

class ParcelKoro:
  
    def __init__(self, name = "No name set", product_weight = 0,):
        self.name = name
        self.product_weight = product_weight
    
    def calculateFee(self,location = None):
        self.location = location
        if self.location == None:
            self.location_charge = 50
        else:
            self.location_charge = 100
    
    def printDetails(self):
        print ("Customer Name:",self.name)
        print ("Product Weight:",self.product_weight)
        if self.product_weight == 0:
            total_fee = 0
        else:
            total_fee = (self.product_weight * 20) + self.location_charge
        print ("Total fee:",total_fee)
    
    
print("**********************")
p1 = ParcelKoro() 
p1.calculateFee()
p1.printDetails()
print("**********************")
p2 = ParcelKoro('Bob The Builder') 
p2.calculateFee()
p2.printDetails()
print("----------------------------") 
p2.product_weight = 15
p2.calculateFee()
p2.printDetails() 
print("**********************")
p3 = ParcelKoro('Dora The Explorer', 10) 
p3.calculateFee('Dhanmondi')
p3.printDetails()
```

```python
#Task_13

class Batsman:
    
    def __init__(self,*info):
        if len(info) == 3:
            self.name = info[0]
            self.runs_scored = info[1]
            self.balls_faced = info[2]
        else:
            self.name = "New Batsman"
            self.runs_scored = info[0]
            self.balls_faced = info[1]
            
    def printCareerStatistics(self):
        print ("Name:",self.name)
        print ("Runs Scored:",self.runs_scored,", Balls Faced:",self.balls_faced)
        
            
    def battingStrikeRate(self):
        x = (self.runs_scored / self.balls_faced) * 100
        return x
    
    def setName(self,setname):
        self.name = setname
        
b1 = Batsman(6101, 7380)
b1.printCareerStatistics() 
print("============================")
b2 = Batsman("Liton Das", 678, 773) 
b2.printCareerStatistics()
print("----------------------------") 
print(b2.battingStrikeRate()) 
print("============================") 
b1.setName("Shakib Al Hasan")
b1.printCareerStatistics() 
print("----------------------------") 
print(b1.battingStrikeRate())
```

```python
#Task_14

class EPL_Team:
    
    def __init__(self, name, song = "No Slogan", title= None):
        self.name = name
        self.song = song
        if title == None:
            self.title = 0
    
    def showClubInfo(self):
        return "Name:{}\nSong:{}\nTotal No of title:{}".format(self.name, self.song, self.title)
    
    def increaseTitle(self):
        self.title += 1
    
    def changeSong (self,changesong):
        self.song = changesong

manu = EPL_Team('Manchester United', 'Glory Glory Man United') 
chelsea = EPL_Team('Chelsea')
print('===================')
print(manu.showClubInfo())
print('##################')
manu.increaseTitle()
print(manu.showClubInfo()) 
print('===================')
print(chelsea.showClubInfo())
chelsea.changeSong('Keep the blue flag flying high')
print(chelsea.showClubInfo())
```

```python
#Task_15
class Account:
    
    def __init__ (self, name = "Default Account", balance = 0.0):
        self.name = name
        self.balance = balance
        
    def details(self):
        return "{}\n{}".format(self.name, self.balance)
    
    def withdraw(self,withdraw):
        self.balance =self.balance - withdraw
        if self.balance < 3070:
            print ("Sorry, Withdraw unsuccessful! The account balance after deducting withdraw amount is equal to or less than minimum.")
        else:
            
            print ("Withdraw successful! New balance is:",self.balance)
            
a1 = Account() 
print(a1.details())
print("------------------------") 
a1.name = "Oliver"
a1.balance = 10000.0 
print(a1.details()) 
print("------------------------") 
a2 = Account("Liam") 
print(a2.details()) 
print("------------------------") 
a3 = Account("Noah",400) 
print(a3.details()) 
print("------------------------")
a1.withdraw(6930); 
print("------------------------")
a2.withdraw(600);
print("------------------------") 
a1.withdraw(6929)
```
